```python
from sympy import *
init_printing(use_unicode=True)
```

```python
x,b,t,mu, sigma = symbols('x b t mu sigma')
```

```python
normal = 1 / sqrt(2 * pi * sigma ** 2) * exp(-0.5 * (((x - mu) / sigma) ** 2))
laplace = 1 / (2 * b) * exp(-(abs(x - t) / b))
laplace_neg = log(1 / (2 * b)) + ((t - x) / b)
laplace_pos = log(1 / (2 * b)) + ((x - t) / b)
```

```python
ce = -normal * log(laplace)
ce
```

```python
ce_int = integrate(ce.subs(t,0), (x, -oo, oo))
```

```python
laplace_neg, laplace_neg.subs(t,0)
```

```python
integrate(normal*laplace_neg.subs(t,0), (x, -oo, 0))+integrate(normal*laplace_pos.subs(t,0), (x, 0, oo))
```
